```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------
# Physical constants (SI units)
# ---------------------------------
G = 6.674e-11
M = 1.989e30
c = 3.0e8

# Mercury orbital parameters
a = 5.79e10          # semi-major axis (m)
e = 0.206            # eccentricity

# Relativistic correction (true value)
delta = 3 * G * M / (a * (1 - e**2) * c**2)

# ---------------------------------
# Angular range: many revolutions
# ---------------------------------
N_orbits = 200       # enough to see precession
phi = np.linspace(0, 2 * np.pi * N_orbits, 200000)

# ---------------------------------
# Newtonian orbit
# ---------------------------------
r_newton = a * (1 - e**2) / (1 + e * np.cos(phi))

# ---------------------------------
# Einstein (GR) orbit
# ---------------------------------
r_gr = a * (1 - e**2) / (1 + e * np.cos((1 - delta) * phi))

# ---------------------------------
# Convert to Cartesian coordinates
# ---------------------------------
xN = r_newton * np.cos(phi)
yN = r_newton * np.sin(phi)

xGR = r_gr * np.cos(phi)
yGR = r_gr * np.sin(phi)

# ---------------------------------
# Plot comparison
# ---------------------------------
plt.figure(figsize=(7, 7))
plt.plot(xN, yN, label="Newtonian orbit")
plt.plot(xGR, yGR, label="Einstein orbit (GR)")
plt.scatter(0, 0, s=80, label="Sun")

plt.axis("equal")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Mercury Orbit: Newtonian vs GR (Perihelion Precession)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Physical constants (SI units)
# -----------------------------
G = 6.674e-11
M = 1.989e30
c = 3.0e8

# -----------------------------
# Mercury orbital parameters
# -----------------------------
a = 5.79e10        # semi-major axis (m)
e = 0.206          # eccentricity

# True relativistic correction
delta = 3 * G * M / (a * (1 - e**2) * c**2)

# Slightly exaggerate delta for visualization
delta_vis = delta * 1e5  # small enough to see perihelion shift clearly

# -----------------------------
# Angular domain: single orbit
# -----------------------------
phi = np.linspace(0, 2*np.pi, 5000)  # one orbit

# -----------------------------
# Newtonian orbit
# -----------------------------
r_newton = a * (1 - e**2) / (1 + e * np.cos(phi))

# -----------------------------
# Einstein (GR) orbit with slightly exaggerated delta
# -----------------------------
r_gr = a * (1 - e**2) / (1 + e * np.cos((1 - delta_vis) * phi))

# -----------------------------
# Convert to Cartesian coordinates
# -----------------------------
xN = r_newton * np.cos(phi)
yN = r_newton * np.sin(phi)

xGR = r_gr * np.cos(phi)
yGR = r_gr * np.sin(phi)

# -----------------------------
# Zoom near perihelion (around x-axis)
# -----------------------------
plt.figure(figsize=(7, 7))
plt.plot(xN, yN, label="Newtonian orbit")
plt.plot(xGR, yGR, label="Einstein orbit (GR, zoomed perihelion)")
plt.scatter(0, 0, s=80, color="orange", label="Sun")
plt.axis("equal")

# Zoom limits around perihelion
plt.xlim([a*(1-e)-1e9, a*(1+e)+1e9])
plt.ylim([-1.5e10, 1.5e10])

plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Mercury Orbit: Zoom near Perihelion")
plt.legend()
plt.show()
```
